### Starting kit on health care data

```python
%matplotlib inline
import numpy as np
from scipy import io
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = io.loadmat('dataMarathon.mat')
```

```python
Z = np.c_[data['data_target'].astype(np.int), data['X']]
```

```python
columns = [u'TARGET'] + [d[0] for d in data['header'].ravel()]
df = pd.DataFrame(Z, columns=columns)
```

```python
df.head()
```

```python
from sklearn.cross_validation import train_test_split

Z = df.values
y = Z[:, 0]
X = Z[:, 1:]

Z_train, Z_test = train_test_split(
    Z, test_size=0.2, random_state=61)
```

```python
train = pd.DataFrame(Z_train, columns=columns)
test = pd.DataFrame(Z_test, columns=columns)
```

```python
train.to_csv("train.csv", index=False, sep=',')
```

```python
test.to_csv("test.csv", index=False, sep=',')
```
